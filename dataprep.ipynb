```python
start_date = "Jan 1, 2022"
end_date = "Dec 31, 2022"
index = pb.date_range(start_date, end_date)
print(f"length in days  of index is {len(index)}")

items_sold = pb.Series(np.random.randint(25, 50, size = len(index)),index = index)
items_sold.plot(figsize =(8,10))
pnp.show()
```
